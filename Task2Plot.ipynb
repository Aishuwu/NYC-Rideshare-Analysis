```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Replace these paths with the actual paths to your CSV files
trips_file = 'output2/trips_per_month/part-00000-755e6663-b629-47a6-806f-25171a8b7214-c000.csv'
profits_file = 'output2/profits_per_month/part-00000-fe1fc471-83e7-4638-9e0e-b5f1ae99e8aa-c000.csv'
earnings_file = 'output2/earnings_per_month/part-00000-8cf4d8c0-f10d-4a5b-bf95-40a829999eeb-c000.csv'

# Load the data into pandas dataframes
df_trips = pd.read_csv(trips_file)
df_profits = pd.read_csv(profits_file)
df_earnings = pd.read_csv(earnings_file)

# Create 'business-month' column
df_trips['business-month'] = df_trips['business'] + '-' + df_trips['month'].astype(str)
df_profits['business-month'] = df_profits['business'] + '-' + df_profits['month'].astype(str)
df_earnings['business-month'] = df_earnings['business'] + '-' + df_earnings['month'].astype(str)

# Function to add data point labels on the bars
def add_data_labels(ax):
    for bar in ax.patches:
        # Get the height of the bar (which represents the data value)
        height = bar.get_height()
        # Place a text label above the bar
        ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.0f}', 
                ha='center', va='bottom')

sns.set_style("whitegrid")

# Plotting number of trips per business per month
plt.figure(figsize=(10, 6))
ax_trips = sns.barplot(data=df_trips, x='business-month', y='count', color="orange")
plt.xticks(rotation=45)
plt.title('Number of Trips per Business per Month')
plt.ylabel('Trip Counts')
plt.xlabel('Business-Month')
add_data_labels(ax_trips)
plt.show()

# Plotting platform profit's per business per month
plt.figure(figsize=(10, 6))
ax_profits = sns.barplot(data=df_profits, x='business-month', y='total_profit', color="blue")
plt.xticks(rotation=45)
plt.title("Platform's Profits per Business per Month")
plt.ylabel('Total Profits')
plt.xlabel('Business-Month')
add_data_labels(ax_profits)
plt.show()

# Plotting driver's earnings per business per month
plt.figure(figsize=(10, 6))
ax_earnings = sns.barplot(data=df_earnings, x='business-month', y='total_earnings', color="green")
plt.xticks(rotation=45)
plt.title("Driver's Earnings per Business per Month")
plt.ylabel('Total Earnings')
plt.xlabel('Business-Month')
add_data_labels(ax_earnings)
plt.show()
```
